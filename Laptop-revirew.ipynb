```python
def categorize_rating(rating):
    if rating > 4:
        return 'Excellent'
    elif 3 < rating <= 4:
        return 'Good'
    elif 2 < rating <= 3:
        return 'Average'
    else:
        return 'Poor'

DS['overall_rating'] = DS['overall_rating'].apply(categorize_rating)
```
