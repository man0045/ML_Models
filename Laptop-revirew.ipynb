```python
def categorize_rating(rating):
    if rating > 4:
        return 'Excellent'
    elif 3 < rating <= 4:
        return 'Good'
    elif 2 < rating <= 3:
        return 'Average'
    else:
        return 'Poor'

DS['overall_rating'] = DS['overall_rating'].apply(categorize_rating)
```

```python
import re
nltk.download('stopword')
nltk.download('punkt_tab')
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
stemmer = PorterStemmer()

DS['sentences'] = DS['review'].apply(nltk.sent_tokenize)
```

```python
import re
corpus = []
for i in range(len(DS['sentences'])):
 new_sent = re.sub(r'[^a-zA-Z]', ' ', DS['review'][i])
 new_sent = new_sent.lower()
 new_sent = new_sent.split()
 review = [stemmer.stem(word) for word in new_sent if not word in set(stopwords.words('english'))]
 new_sent = ' '.join(review)
 corpus.append(new_sent)
print(corpus)
print(len(corpus))
```

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer
stemmer = PorterStemmer()
stemmed = [stemmer.stem(word) for word in corpus]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform([" ".join(stemmed)]).toarray()
```
